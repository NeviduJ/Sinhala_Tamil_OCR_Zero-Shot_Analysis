```python
from jiwer import wer, cer
```

```python
import pandas as pd
import re
```

```python
df = pd.read_csv("subasa_selenium_ocr_results.csv")
```

```python
df.head(30)
```

```python
def remove_non_sinhala(text):
     if isinstance(text, str):
        return ''.join(char for char in text if re.match(r'[\u0D80-\u0DFF\s]', char))
     else:
        return ''  # or some other appropriate default value
```

```python
df['reference'] = df['reference'].apply(remove_non_sinhala)
df['generated_subasa'] = df['generated_subasa'].apply(remove_non_sinhala)
```

```python
df.head(30)
```

```python
agg_error_wer = 0
agg_error_cer = 0
for i in range(len(df)):
    record = df.iloc[i]
    try:
        error_wer = wer(record['reference'], record['generated_subasa'])
        error_cer = cer(record['reference'], record['generated_subasa'])
    except:
        print(record)
        print(i)
    agg_error_wer += error_wer
    agg_error_cer += error_cer
```

```python
print(f"Word Error Rate (WER): {agg_error_wer/ len(df)}")
```

```python
print(f"Character Error Rate (CER): {agg_error_cer/ len(df)}")
```
